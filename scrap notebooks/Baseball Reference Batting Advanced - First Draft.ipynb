```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
year = 2017
url = "https://www.baseball-reference.com/leagues/MLB/2017-standard-batting.shtml"
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page,"html.parser")
```

```python
site = soup.findAll("tbody")
batting_table = soup.find('div', class_='placeholder').next_sibling.next_sibling
batting_table = BeautifulSoup(batting_table,'lxml')
```

```python
players = batting_table.find_all('tr')
```

```python
player_dict={}
line_fail=0
for p in players[1:1511]: #num of hitters in 2017
    td = p.find_all('td')
    if  len(td)>1:
        if td[0].find('a') is None:
            break
        else:
            player=td[0].find('a').text #this gets his full name
            player = player.replace(u'\xa0', u' ') #replaces the unicode space
            player_dict[player] = [i.text for i in td[1:]]
    line_fail+=1
    
#failing at 1511 - may be the end of the data
```

```python
print(player_dict['Yoenis Cespedes'])
```

player_dict is now a dictionary of the 2017 MLB basic hitting stats
